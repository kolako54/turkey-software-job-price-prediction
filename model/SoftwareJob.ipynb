```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
```

```python
len(y)
```

```python
X = df.iloc[:, :-1]
#y = df.iloc[:, -1]
```

```python
len(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
model = CatBoostRegressor(iterations=1000, learning_rate=0.07, depth=7, eval_metric='MAE', random_state=42)
```

```python
X.columns
```

```python
model.fit(X_train, y_train, [0,1,2,3,4,5], eval_set=(X_test, y_test))
```

```python
model.score(X_test, y_test)
```
